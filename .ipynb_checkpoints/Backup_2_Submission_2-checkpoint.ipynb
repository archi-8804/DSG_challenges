```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
%matplotlib inline
# df_1= pd.read_csv('a1_raw.csv')
df_2= pd.read_csv('a1_va3.csv')
# df_3=pd.read_csv('a2_raw.csv')
# df_4= pd.read_csv('a3_raw.csv')
df_5= pd.read_csv('a3_va3.csv')
# df_6=pd.read_csv('b1_raw.csv')
df_7= pd.read_csv('b1_va3.csv')
# df_8=pd.read_csv('b3_raw.csv')
df_9= pd.read_csv('b3_va3.csv')
# df_10=pd.read_csv('c1_raw.csv')
df_11= pd.read_csv('c1_va3.csv')
# df_12=pd.read_csv('c3_raw.csv')
df_13= pd.read_csv('c3_va3.csv')
new_testing_data=pd.read_csv('test.csv')

sample_sub_df=pd.read_csv("sample_submission.csv")

df_2.shape

df_5.shape

df_7.shape

df_9.shape

df_11.shape

df_13.shape
new_testing_data.shape

df_2.info()

df_5.info()

df_7.info()

df_9.info()

df_11.info()

df_13.info()
new_testing_data.info()
df_9.isnull().any()
df_2
print(df_2.describe())

null_values_in_columns = new_testing_data.isnull().any()
print(null_values_in_columns)

list_of_dataframes = [df_2,  df_5, df_7,  df_9,  df_11, df_13]
# concatenated_df = pd.concat(list_of_dataframes, ignore_index=True)
concatenated_df = pd.concat(list_of_dataframes, axis=0, ignore_index=True)

concatenated_df.shape

concatenated_df

concatenated_df.isnull().any()
X = concatenated_df.iloc[:, :-1]
y = concatenated_df.iloc[:, -1]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
X_test = new_testing_data
y_test = sample_sub_df.iloc[:, -1]  
y_test_encoded = label_encoder.transform(y_test)
y_pred_test = model.predict(X_test)

y_pred_test
y_pred_test_original = label_encoder.inverse_transform(y_pred_test)
y_pred_test_original

df_results=pd.DataFrame({"Phase":y_pred_test_original})
df_results
df_results.to_csv("DSG_Archi_Submission_2.csv")
```
